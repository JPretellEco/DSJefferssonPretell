```python
# Librer√¨as que se utilizar√†n en este caso

import numpy as np
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
# Como el archivo es un .ods, se debe de leer con la librer√¨a pandas
# Se lee el archivo con la funci√≤n read_csv

ruta = "./datasets/CasoPelican.ods"

# Se lee el archivo con la funci√≤n read_csv

df = pd.read_excel(ruta,engine = 'odf')

tiendaPelican = df.iloc[:,range(1,6)].rename(df["Cliente"].to_dict())

```

```python
# Haremos un resumen de los 5 primeros n√πmeros

tiendaPelican.describe(include='all')
```

### ¬øCu√†l es el g√®nero que compr√≤ m√†s?

```python
gnro = tiendaPelican.loc[:,["Articulos","Genero","Tipo de Cliente"]].groupby(["Tipo de Cliente","Genero"]).sum().plot.bar()
plt.show()
```

### ¬øCu√†l fue el ticket promedio y cantidad promedio de art√¨culos vendidos?

```python
tkprom = tiendaPelican.loc[:,["Articulos","Ventas_Netas","Genero","Tipo de Cliente","M√®todo de pago"]].groupby(["Tipo de Cliente","M√®todo de pago","Genero"]).mean()
tkprom
```

### ¬øCu√†l fue el m√®todo de pago m√†s utilizada?

```python
ax = tiendaPelican["M√®todo de pago"].value_counts().plot.bar()
ax.set_title("M√®todos de pagos m√†s utilizados")
ax.set_ylabel("Cantidad de pagos")
```

### Por √πltimo,¬øqu√® tipo de cliente asiti√≤ m√†s?

```python
ax = tiendaPelican["Tipo de Cliente"].value_counts().plot.bar()
ax.set_title("Tipo de clientes que m√†s asistieron")
ax.set_ylabel("Cantidad de clientes")
```
