```python
from io import StringIO

import pandas as pd

csv_file = StringIO(""" Pelicula,Ventas_Brutas_Estreno,Ventas_Brutas_Totales,Nro_Salas,  Semana_60mejores
    Coach Carter ,29.17, 67.25, 2574, 16
    Ladies in Lavender ,0.15, 6.65 ,119 ,22
    Batman Begins ,48.75, 205.28 ,3858 ,18
    Unleashed, 10.90 ,24.47,1962 ,8
    Pretty Persuasion, 0.06 ,0.23, 24, 4
    Fever Pitch ,12.40 ,42.01, 3275 ,14
    Harry Potter and the Goblet of Fire ,102.69, 287.18, 3858, 13
    Monster-in-Law, 23.11 ,82.89 ,3424 ,16
    White Noise ,24.11, 55.85, 2279, 7
    Mr. and Mrs. Smith, 50.34, 186.22, 3451, 21""")
data3 = pd.read_csv(csv_file)
data3
```

### Pel√¨culas m√†s vendidas en estreno

```python
import matplotlib.pyplot as plt

ax = data3.loc[:,["Ventas_Brutas_Estreno"]].groupby(data3[" Pelicula"]).mean().sort_values(by= "Ventas_Brutas_Estreno",ascending=False).plot.bar()
ax.set_title("Ranking de las pel√¨culas con mayores ventas en sus estrenos")
ax.set_ylabel("Ventas")
plt.show()


```

### ¬øQu√® pel√¨culas ocuparon m√†s salas? 


```python
ax2 = data3.loc[:,["Nro_Salas"]].groupby(by = data3[" Pelicula"]).mean().sort_values(by="Nro_Salas",ascending=False ).plot.bar(color="Green")
ax2.set_title("Peliculas con m√†s salas oczupadas")
ax2.set_ylabel("Salas ocupadas")
plt.show()
```

### ¬øQu√® pel√¨culas m√†s duraron en el ranking de las 60 mejores pel√¨culas?

```python
ax3 = data3.loc[:,"  Semana_60mejores"].groupby(by = data3[" Pelicula"]).mean().sort_values(ascending=False).plot.bar()
ax3.set_title("Las pel√¨culas con m√†s tiempo en el ranking de las 60 mejores pel√¨culas")
ax3.set_ylabel("Semanas")
plt.show()
```

### ¬øSer√† que las ventas de los estrenos tienen relaci√≤n con las ventas brutas?

```python
import matplotlib.pyplot as plt 

fig= plt.figure(figsize=(6,6))

ax4 = fig.add_subplot(111)

ax4.scatter(data3["Ventas_Brutas_Totales"],data3["Ventas_Brutas_Estreno"])
ax4.set_title("Relaci√≤n entre la ventas brutas totales y las estrenos")
ax4.set_ylabel("Ventas brutas totales")
ax4.set_xlabel("Ventas totales de los estrenos")
plt.show()
```
