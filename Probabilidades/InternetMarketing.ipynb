Imagina que tienes una tienda en línea llamada **"SportStyle"** que vende ropa deportiva, equipos de golf, y accesorios deportivos de última tendencia. La tienda quiere mejorar la experiencia de sus clientes mostrando ofertas personalizadas según el género del visitante, para lograr una mayor conversión de ventas. 🛍️💻

En la tienda **SportStyle**, se sabe que aproximadamente el 60% de los visitantes son hombres (🏌️‍♂️) y el 40% son mujeres (👩‍🦱). Pero, **SportStyle** también desea utilizar información extra sobre los hábitos de navegación de los usuarios para hacer recomendaciones más acertadas.

Aquí es donde entra el poder del **modelo Bayesiano**. Se sabe que si un visitante de la tienda web previamente navegó por el sitio web de **"FashionTrends"** (una tienda en línea de ropa de moda), es **tres veces más probable que ese sitio haya sido visitado por mujeres** en comparación con hombres.

### Requerimientos:
1. **Probabilidad previa**: ¿Cuál es la probabilidad de que el siguiente visitante que entre a la tienda **SportStyle** sea mujer? 
2. **Probabilidad revisada**: Supón que el visitante actual de **SportStyle** visitó previamente **FashionTrends**. ¿Cuál es la probabilidad revisada de que el visitante sea mujer? ¿Deberías mostrar la oferta para mujeres o para hombres? 🛍️🤔

### Datos:
- La **probabilidad previa** de que el visitante sea mujer es 40% (es decir, P(mujer) = 0.4).
- La probabilidad de que una mujer visite **FashionTrends** es 3 veces más probable que la de un hombre.
  - Esto se expresa como P(FashionTrends|mujer) = 3 * P(FashionTrends|hombre).

# Probabilidad de que el siguiente visitante sea mujer

```python
PH = 0.6
PM = 0.4

print(f"Sabemos que por conocimientos previos, \n la probabilidad de que el siguiente visitante al sitio web sea una mujer es {PM}")
```

# Cuàl es la probabilidad de que visitante de FashionTrend visite la pàgina de Dillard

```python

from sympy import symbols,Eq,solve

p = symbols('p') # Definimos la variable 

PMD = 3*p # La probabilidad de que el visitante de Dillar sea mujer es 3 veces màs que la de un hombre
PHD = p # Probabilidad de que el visitante de Dillar sea hombre

PD = PMD*PM + PHD*PH # Probabilidad de que se visite la pàgina de Dillard

PD

print("La probabilidad de que un visiten la pàgina de Dillard es: ",PD)
```

# Probabilidad de que el visitante de Dillar sea Mujer

### Teorema de Bayes Aplicado al Problema




```python
PDM = (PMD*PM)/PD # Probabilidad de que el visitante de Dillar sea mujer dado que se visitò la pàgina de Dillard
print("La probabilidad de que el visitante de Dillar sea mujer dado la probabilidad de que  visite la pàgina de Dillard  \n es: ",round(PDM,4))

# Para redondear un float a 3 decimales se puede usar el método que no sea round 
```
