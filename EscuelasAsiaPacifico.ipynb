### ¬øQui√®n gasta m√†s un alumno extranjero o local?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que 'escuelas' es tu escuelasFrame
# Crea la figura y los ejes para los diagramas de caja
fig, axes = plt.subplots(1, 2, figsize=(10, 6))

# Dibuja el diagrama de caja para "Colegiatura_estudianetes_locales($)"
escuelas["Colegiatura_estudianetes_locales($)"].plot.box(ax=axes[0], title="Locales")

# Dibuja el diagrama de caja para "Colegiatura_estudiantes_fuera($)"
escuelas["Colegiatura_estudiantes_fuera($)"].plot.box(ax=axes[1], title="Extranjeros")

# Configura los t√≠tulos
axes[0].set_title("Estudiantes Locales")
axes[1].set_title("Estudiantes Extranjeros")

# Configura el layout
plt.tight_layout()

# Muestra la figura
plt.show()
```

### Saber ingl√®s,¬øhace ganar m√†s?

```python
tabla2 = escuelas.loc[:,["Salario_inicial","Colegiatura_estudiantes_fuera($)","Colegiatura_estudianetes_locales($)"]].corr()
tabla2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear el heatmap con correcciones
ax = sns.heatmap(
    tabla2,
    vmin=-1,  # Correlaci√≥n m√≠nima t√≠pica
    vmax=1,   # Correlaci√≥n m√°xima t√≠pica
    cmap=sns.diverging_palette(20, 220, as_cmap=True),  # Paleta de colores
    annot=True  # Para mostrar los valores en el mapa de calor
)

ax.set_title("Correlaci√≤n entre salarios y costo de colegiaturas")
plt.show()
```
